```python
#Verilerin görselleştirilmesi
def visualize_random_images(dataset, title):
    plt.figure(figsize=(12,10))
    
    for i in range(10):
        random_image_path = random.choice(dataset)
        
        img = mpimg.imread(random_image_path)
        plt.subplot(1, 10, i+1)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        plt.title(title)
    plt.show()
```

```python
print("Train Data")
visualize_random_images(train_normal, 'Normal')
visualize_random_images(train_pneumonia, 'Pnömoni')
```

```python
print("Test Data")
visualize_random_images(test_normal, 'Normal')
visualize_random_images(test_pneumonia, 'Pnömoni')
```

```python
print("Validation Data")
visualize_random_images(val_normal, 'Normal')
visualize_random_images(val_pneumonia, 'Pnömoni')
```
